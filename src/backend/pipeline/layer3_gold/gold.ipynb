```python
import os
import sys

import pandas as pd

current_dir = os.getcwd()
parent_dir = os.path.abspath(os.path.join(current_dir, "..", "..", ".."))
data_dir = os.path.abspath(os.path.join(current_dir, "..", "..", "..", "..", "data"))
sys.path.append(parent_dir)
```

```python
path_parquet = os.path.join(data_dir, "silver", "breweries.parquet")
df = pd.read_parquet(path_parquet, "pyarrow")
```

```python
df_aggregate = (
    df
    .groupby(["brewery_type", "country", "state"], observed=True)["id"] # Apenas as combinações que ocorreram '>0'
    .count()
    .reset_index(name="quantidade")
    )\
    .sort_values(['quantidade'], ascending=False)

df_aggregate.head(10) # top10
```

```python
df_aggregate.to_parquet( # save in layer gold
    path=os.path.join(
        data_dir, "gold", "breweries.parquet"
    )
)
```
